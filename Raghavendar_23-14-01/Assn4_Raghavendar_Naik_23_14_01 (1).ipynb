**6 Replace the NaN and String entries in Num_Bath with previous entry**

---



```python
pd.set_option('mode.chained_assignment', None)
hd['NUM_BATH']=pd.to_numeric(hd['NUM_BATH'], errors='coerce')
hd['NUM_BATH']=hd['NUM_BATH'].fillna(method='ffill')
hd
```

**7. Replace the empty values in “SQ_FT” with the mean of the all entries.**

```python
hd['SQ_FT'] = pd.to_numeric(hd['SQ_FT'], errors='coerce')
hd['SQ_FT']=hd['SQ_FT'].fillna(value=hd.SQ_FT.mean())
hd
```

16. Display the state having highest average price per square foot of housing area. Hint: Average of
(TotalPrice/Sq_Ft)

```python
hd_temp=hd.groupby(['STATE'])[['PRICE', 'SQ_FT']].sum()
hd_temp['avg_price']=hd_temp["PRICE"]/hd_temp['SQ_FT']
hd_temp = hd_temp.sort_values(by=['avg_price'],ascending=False)
```
